```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\SPARTAN PC\Desktop\Dataton\Datasets\FAOSTAT_emisiones_mx.csv")
```

```python
df.head()
```

```python
df = df.drop('Note', axis=1)
```

```python
df.head()
```

```python
df.groupby('Year Code')['Value']
```

```python
plt.figure(figsize=(8, 6))
plt.bar(df['Year'], df['Value'], color='skyblue') #Barras
plt.plot(df['Year'], df['Value'], color='green') #Linea 
plt.xlabel('Año')
plt.ylabel('Emisiones en kilo toneladas')
plt.title('Emisiones de metano por cria de ganado en México (1960-2022)')
plt.show()
```
